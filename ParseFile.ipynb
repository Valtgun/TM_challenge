```python
newdata = dataset.copy()
```

```python
newdata["dz_gads"] = newdata['Dzimšanas datums'].astype(str).str[-2:]
#TODO:
# Prasijuma apmers -> percentiles binned
newdata["ien_datums"] = pd.to_datetime(newdata['Ienākšanas datums'])
newdata["izsk_datums"] = pd.to_datetime(newdata['Izskatīšanas datums'])
newdata["menesi"] = (newdata["izsk_datums"].dt.year - newdata["ien_datums"].dt.year) * 12 + (newdata["izsk_datums"].dt.month - newdata["ien_datums"].dt.month)
#newdata.head()
```

```python
# Convert Tiesa
newdata["T32"] = np.where(newdata['Tiesas ID (courtID)']==32, 1, 0)
newdata["T26"] = np.where(newdata['Tiesas ID (courtID)']==26, 1, 0)
newdata["T279"] = np.where(newdata['Tiesas ID (courtID)']==279, 1, 0)
newdata["T22"] = np.where(newdata['Tiesas ID (courtID)']==22, 1, 0)
newdata["T29"] = np.where(newdata['Tiesas ID (courtID)']==29, 1, 0)
newdata["T34"] = np.where(newdata['Tiesas ID (courtID)']==34, 1, 0)
newdata["T273"] = np.where(newdata['Tiesas ID (courtID)']==273, 1, 0)
newdata["T27"] = np.where(newdata['Tiesas ID (courtID)']==27, 1, 0)
newdata["T23"] = np.where(newdata['Tiesas ID (courtID)']==23, 1, 0)
newdata["T28"] = np.where(newdata['Tiesas ID (courtID)']==28, 1, 0)
newdata["T31"] = np.where(newdata['Tiesas ID (courtID)']==31, 1, 0)
newdata["T19"] = np.where(newdata['Tiesas ID (courtID)']==19, 1, 0)
newdata["T6"] = np.where(newdata['Tiesas ID (courtID)']==6, 1, 0)
newdata["T30"] = np.where(newdata['Tiesas ID (courtID)']==30, 1, 0)
newdata["T4"] = np.where(newdata['Tiesas ID (courtID)']==4, 1, 0)
newdata["T1"] = np.where(newdata['Tiesas ID (courtID)']==1, 1, 0)
newdata["T8"] = np.where(newdata['Tiesas ID (courtID)']==8, 1, 0)
newdata["T5"] = np.where(newdata['Tiesas ID (courtID)']==5, 1, 0)
newdata["T39"] = np.where(newdata['Tiesas ID (courtID)']==39, 1, 0)
newdata["T38"] = np.where(newdata['Tiesas ID (courtID)']==38, 1, 0)
newdata["T2"] = np.where(newdata['Tiesas ID (courtID)']==2, 1, 0)
newdata["T16"] = np.where(newdata['Tiesas ID (courtID)']==16, 1, 0)
newdata["T18"] = np.where(newdata['Tiesas ID (courtID)']==18, 1, 0)
newdata["T37"] = np.where(newdata['Tiesas ID (courtID)']==37, 1, 0)
newdata["T10"] = np.where(newdata['Tiesas ID (courtID)']==10, 1, 0)
newdata["T7"] = np.where(newdata['Tiesas ID (courtID)']==7, 1, 0)
newdata["T40"] = np.where(newdata['Tiesas ID (courtID)']==40, 1, 0)
newdata["T25"] = np.where(newdata['Tiesas ID (courtID)']==25, 1, 0)
newdata["T33"] = np.where(newdata['Tiesas ID (courtID)']==33, 1, 0)
newdata["T284"] = np.where(newdata['Tiesas ID (courtID)']==284, 1, 0)

newdata.head()
```

```python
# Convert Ienākšanas kārtība
newdata["IK_Pirmo"] = np.where(newdata['Ienākšanas kārtība']=='Pirmo reizi', 1, 0)
newdata["IK_CPL"] = np.where(newdata['Ienākšanas kārtība'].astype(str).str[:3]=='CPL', 1, 0)
newdata["IK_Apv"] = np.where(newdata['Ienākšanas kārtība'].astype(str).str[:5]=='Sakar', 1, 0)
newdata["IK_Piekr"] = np.where(newdata['Ienākšanas kārtība']=='Pēc piekritības no citas tiesas', 1, 0)
newdata["IK_Atcel"] = np.where(newdata['Ienākšanas kārtība']=='Pēc lēmuma atcelšanas', 1, 0)
newdata["IK_Atdal"] = np.where(newdata['Ienākšanas kārtība']=='Lietas atdalīšana', 1, 0)
newdata.head()
```

```python
# Convert Būtība
newdata["B_upper"] = newdata['Būtība'].str.upper()
newdata["B_maza"] = np.where(newdata['B_upper'].str.contains('MAZA'), 1, 0)
newdata["B_zaud"] = np.where(newdata['B_upper'].str.contains('ZAUD'), 1, 0)
newdata["B_nodro"] = np.where(newdata['B_upper'].str.contains('NODRO'), 1, 0)
newdata["B_parada"] = np.where(newdata['B_upper'].str.contains('PARĀ'), 1, 0)
newdata["B_proc"] = np.where(newdata['B_upper'].str.contains('PROC'), 1, 0)
newdata["B_izli"] = np.where(newdata['B_upper'].str.contains('IZLIK'), 1, 0)
newdata["B_izbeig"] = np.where(newdata['B_upper'].str.contains('IZBEIG'), 1, 0)
newdata["B_apdro"] = np.where(newdata['B_upper'].str.contains('APDRO'), 1, 0)
newdata["B_other"] = np.where((newdata['B_maza']+newdata["B_zaud"]+newdata["B_nodro"]+newdata["B_parada"]+newdata["B_proc"]+newdata["B_izli"]+newdata["B_izbeig"]+newdata["B_apdro"])==0, 1, 0)
newdata.head()
```

```python
# Prasijuma apmers -> percentiles binned
newdata['Prasījuma apmērs'] = np.where(newdata['Prasījuma apmērs']==0, 1, newdata['Prasījuma apmērs'])
newdata["Pras_log"] = np.log10(newdata['Prasījuma apmērs'])
newdata.head()
```

```python
newdata.describe()
```

```python
newdata["Pras_log"] = np.where(newdata["Pras_log"]<0, 0, newdata["Pras_log"])
newdata["Pras_log"] = np.where(newdata["Pras_log"]>5, 5, newdata["Pras_log"])
newdata.describe()
```

```python
# Normalization
newdata["Pras_log"] = newdata["Pras_log"]/5
```

```python
newdata.describe()
```

```python
newdata["dz_gads_norm"] = (120-newdata["dz_gads"].astype(int)*2+65)/100
newdata["atb_f_n"] = newdata["atb_fiz"]/20
newdata["atb_j_n"] = newdata["atb_jur"]/5
newdata["pra_f_n"] = newdata["pras_fiz"]/5
newdata["pra_j_n"] = newdata["pras_jur"]/5
newdata["noz_n"] = newdata["Nozīmētās sēdes"]/30
newdata["not_n"] = newdata["Notikušās sēdes"]/15
newdata["naso_n"] = newdata["naudaS_sods"]/5
newdata["arz_n"] = newdata["dzivo_arzemes"]/5

newdata["atb_f_n"] = np.where(newdata["atb_f_n"]>1, 1, newdata["atb_f_n"])
newdata["atb_j_n"] = np.where(newdata["atb_j_n"]>1, 1, newdata["atb_j_n"])
newdata["pra_f_n"] = np.where(newdata["pra_f_n"]>1, 1, newdata["pra_f_n"])
newdata["pra_j_n"] = np.where(newdata["pra_j_n"]>1, 1, newdata["pra_j_n"])
newdata["noz_n"] = np.where(newdata["noz_n"]>1, 1, newdata["noz_n"])
newdata["not_n"] = np.where(newdata["not_n"]>1, 1, newdata["not_n"])
newdata["naso_n"] = np.where(newdata["naso_n"]>1, 1, newdata["naso_n"])
newdata["arz_n"] = np.where(newdata["arz_n"]>1, 1, newdata["arz_n"])
```

```python
newdata.describe()
```

```python
# normalize start date
newdata["ien_gads_n"] = (newdata["ien_datums"].dt.year - 2000)*12 + newdata["ien_datums"].dt.month
```

```python
writer = pd.ExcelWriter('output.xlsx')
newdata.to_excel(writer,'Sheet1')
writer.save()
```
